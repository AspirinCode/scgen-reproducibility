```python
import sklearn
import scanpy.api as sc
sc.logging.print_versions()
```

original data 

```python
adata_orig = sc.read("./data/pancreas.h5ad")
```

```python
sc.tl.pca(adata_orig, svd_solver='arpack')
adata_orig.obsm['X_pca'] *= -1  # multiply by -1 to match Seurat
sc.pp.neighbors(adata_orig)
sc.tl.umap(adata_orig)
```

```python
sc.pl.umap(adata_orig, color="batch", legend_loc='none', frameon=False, title=" ")
sc.pl.umap(adata_orig, color="celltype", legend_loc='none', frameon=False,  title=" ")
```

```python
sc.tl.pca(adata_orig, svd_solver='arpack')
adata_orig.obsm['X_pca'] *= -1  # multiply by -1 to match Seurat
X = adata_orig.obsm["X_pca"]
labels = adata_orig.obs["batch"].tolist()
print(f"average silhouette_score for original data :{sklearn.metrics.silhouette_score(X,labels)}")
```
